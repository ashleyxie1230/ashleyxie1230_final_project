```python
%matplotlib inline
import folium
import geopandas as gpd
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
crimerate = pd.read_csv("crime rate.csv")
```

```python
crimerate = crimerate.rename(columns={"YEAR":"State"})
x = []
for i in range(2006,2015):
    y = crimerate[["State",str(i)]]
    y = y.rename(columns={str(i):"Crime Rate"})
    y["Year"] = i
    x.append(y)
```

```python
allcrime = x[0]
for i in range(1,len(x)):
    allcrime = pd.concat([allcrime,x[i]], axis=0)
```

```python
y = []
for i in range(2006,2015):
    poverty = pd.read_csv("poverty"+str(i)+".csv")
    poverty["Year"] = i
    poverty = poverty.rename(columns={"STATE":"State"})
    y.append(poverty)
```

```python
allpoverty = y[0]
for i in range(1,len(y)):
    allpoverty = pd.concat([allpoverty,y[i]], axis=0)
```

```python
merged = pd.merge(allcrime,allpoverty)
```

```python
merged = merged.rename(columns={"Percent":"Poverty Percent"})
merged.to_csv("merged.csv", encoding='utf-8')
merged
```

```python
npMatrix = np.matrix(merged[["Poverty Percent","Crime Rate"]])

from sklearn.feature_selection import chi2
x,y = npMatrix[:,0], npMatrix[:,1]
model = LinearRegression().fit(x,y)
m,b = model.coef_[0],model.intercept_
pvalue = model.
print (pvalue)

title = "LinearRegression Formula: y = {0}x + {1}".format(m,b)

plt.scatter(x,y,color='blue')
plt.plot([0,25],[b,m*25+b],'r')
plt.title(title)
plt.xlabel('Poverty Percent', fontsize = 15)
plt.ylabel('Crime Rate', fontsize = 15)
plt.savefig("regression.png")
```

```python
mean = merged.groupby('State')[['Crime Rate','Poverty Percent']].mean()
mean
```

```python
geo_df = gpd.read_file("cb_2015_us_state_20m.shp")
geo_df = geo_df.rename(columns={"NAME":"State"})
geo_df.set_index(geo_df["State"], inplace = True)
mean["Crime Rate"] = mean["Crime Rate"].astype(float)
geo_merge = geo_df.join(mean, how="inner")
geo_merge["Crime Rate"] = geo_merge["Crime Rate"].astype(float)
m = folium.Map([39.828175, -98.5795], 
               tiles='cartodbpositron', 
               zoom_start=4, max_zoom=14, min_zoom=4)

ft = "Crime Rate"
cmap = folium.colormap.linear.BuPu.scale(geo_merge[ft].min(), geo_merge[ft].max())


folium.GeoJson(geo_merge,
               style_function=lambda feature: {
                'fillColor': cmap(feature['properties'][ft]),
                'fillOpacity' : 0.6,
                'weight' : 2, 'color' : 'black'
               }).add_to(m)

cmap.caption = 'Average Crime Rate 2006-2014'
cmap.add_to(m)

m.save("crime_rate.html")
m
```
